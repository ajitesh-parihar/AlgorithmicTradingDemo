```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from datetime import datetime
```

```python
df = pd.read_csv('SP500_Historical.csv',delimiter=',')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
plt.figure(figsize=(20,5))
sns.set(style='ticks')
sns.lineplot(data=df,x='date',y='low',label='Low Price point at date', linewidth=1, color='red')
sns.lineplot(data=df,x='date',y='high',label='High Price point at date', linewidth=1, color='blue')
plt.title('Day High Values of S&P500 Index')
plt.ylabel('Price (USD)')
plt.xlabel('Date')
# plt.xticks(rotation=90)

plt.show()
```
